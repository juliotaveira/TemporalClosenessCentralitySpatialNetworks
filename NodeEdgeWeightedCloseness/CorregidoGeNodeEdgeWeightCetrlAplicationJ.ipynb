```python
import pandas as pd

# Assuming the file is in the specified path
file_path = '/content/drive/MyDrive/DatosToUPE/todos_dataframes.xlsx'

try:
    # Read all sheets into a dictionary of DataFrames
    xls = pd.ExcelFile(file_path)
    all_dfs = {}
    for sheet_name in xls.sheet_names:
        all_dfs[sheet_name] = pd.read_excel(xls, sheet_name)

    graph_dict = {}

    for sheet_name, df in all_dfs.items():
        # Ensure the DataFrame is square (symmetric adjacency matrix)
        if df.shape[0] != df.shape[1]:
            print(f"Warning: DataFrame '{sheet_name}' is not square. Skipping.")
            continue

        # Set column names as index (row names)
        df.index = df.columns

        # Extract edges and weights
        edges_with_weights = {}
        for i in range(len(df.index)):
            for j in range(i + 1, len(df.columns)):  # Iterate through upper triangle
                weight = df.iloc[i, j]

                # Convert negative weights to positive by taking the absolute value
                if pd.notna(weight) and weight != 0:
                    edges_with_weights[(df.index[i], df.columns[j])] = abs(weight)


        graph_dict[sheet_name] = edges_with_weights

except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python

# Print dimensions and properties of graph_dict
for sheet_name, edges_with_weights in graph_dict.items():
    print(f"Graph: {sheet_name}")
    print(f"Number of edges: {len(edges_with_weights)}")
    print(f"Edges with weights: {edges_with_weights}")
    print("-" * 30)
```

```python
import random

def plot_random_edge_weights(graph_dict):
    # Get a list of all unique edges across all time points
    all_edges = set()
    for edges_with_weights in graph_dict.values():
        all_edges.update(edges_with_weights.keys())

    if not all_edges:
        print("No edges found in the graph data.")
        return

    # Choose a random edge
    random_edge = random.choice(list(all_edges))
    source, target = random_edge

    print(f"Selected random edge: ({source}, {target})")

    # Extract weights for the chosen edge over time
    weights_over_time = []
    sheet_names = list(graph_dict.keys())

    for sheet_name in sheet_names:
        edges_with_weights = graph_dict[sheet_name]
        weight = edges_with_weights.get(random_edge, 0)  # Get weight, default to 0 if edge doesn't exist
        weights_over_time.append(weight)

    # Prepare labels for the x-axis
    time_labels = [f"{i+1}hr" for i in range(len(sheet_names))]

    # Plot the weights over time
    plt.figure(figsize=(12, 6))
    plt.plot(time_labels, weights_over_time, marker='o', linestyle='-')
    plt.xlabel("Time")
    plt.ylabel("Weight")
    plt.title(f"Weight of Edge ({source}, {target}) Over Time")
    plt.grid(False)
    plt.show()

# Assuming graph_dict is already defined from the previous code
# Call the function to plot the random edge weights
plot_random_edge_weights(graph_dict)
```

## Node Weight Edge Centrality(Full Graph Temporal Metric)

```python
def calculate_temporal_weighted_node_closeness(graph, graph_dict, time_interval):
    """
    Calculates and sums the weighted node closeness centrality for a given time interval,
    using edge flows as node weights for the source node of each edge. The shortest
    paths are calculated based on the original graph's edge weights (fixed).

    Args:
        graph: The base graph object with fixed edge weights (e.g., distance).
        graph_dict: A dictionary where keys are time steps (e.g., 'Time_1', 'Time_2')
                    and values are dictionaries representing edge weights (flow) at that time step: {(source, target): flow}.
                    This data is used ONLY for calculating the temporal node weights.
        time_interval: A tuple (start_time, end_time) representing the time interval
                       (inclusive) for which to calculate the sum of centralities. The values in the tuple
                       should correspond to the *index* of the time step (0-based), not the 'Time_x' string.

    Returns:
        A dictionary where keys are nodes and values are the sum of their
        temporal weighted closeness centralities normalized by the number of time steps
        in the interval, and a dictionary `weight_n_t_history` storing node weights
        for each time step.
    """
    start_time, end_time = time_interval
    if start_time < 0 or end_time >= len(graph_dict) or start_time > end_time:
        print("Invalid time interval.")
        return {}, {}

    all_nodes = list(graph.nodes())
    temporal_closeness_sum = {node: 0.0 for node in all_nodes}
    num_time_steps = end_time - start_time + 1
    weight_n_t_history = {node: [] for node in all_nodes} # Dictionary to store node weights over time

    # Pre-calculate shortest paths based on the fixed edge weights of the original graph
    # Check for negative edge weights in the base graph before calculating shortest paths
    for u, v, data in graph.edges(data=True):
        if 'weight' in data and data['weight'] < 0:
            raise ValueError("Graph contains negative edge weights. Dijkstra's algorithm cannot be used.")
        #elif 'weight' not in data:
            # Add a default weight if not present, e.g., 1 for unweighted edges
             #graph[u][v]['weight'] = 1 # Assign a default weight

    # Calculate all-pairs shortest paths only once
    try:
        # Using all_pairs_dijkstra_path_length for weighted graphs
        all_pairs_shortest_paths = dict(nx.all_pairs_dijkstra_path_length(graph, weight='weight'))
    except ValueError as e:
        print(f"Error calculating shortest paths: {e}")
        return {}, {}


    for t in range(start_time, end_time + 1):
        # Access the correct time step using the 'Time_x' key
        time_step_key = f"Time_{t+1}"
        if time_step_key not in graph_dict:
            print(f"Warning: Time step '{time_step_key}' not found in graph_dict. Skipping.")
            continue

        edge_weights_at_t = graph_dict[time_step_key]

        # Create node weights based on outgoing edge flow at time t (from graph_dict)
        weight_n_t = {node: 0 for node in all_nodes}
        for (u, v), flow in edge_weights_at_t.items():
            if u in weight_n_t:
                weight_n_t[u] += flow

        # Store the node weights for the current time step
        for node in all_nodes:
            weight_n_t_history[node].append(weight_n_t.get(node, 0))

        # Calculate weighted closeness for the current time step using the PRE-CALCULATED shortest paths
        current_closeness = {}
        total_weight_t = sum(weight_n_t.values())

        if total_weight_t > 0:
            for u in all_nodes:
                fu = weight_n_t.get(u, 0)
                sum_fv_div_duv = 0

                if u in all_pairs_shortest_paths:
                    lengths_from_u = all_pairs_shortest_paths[u]
                    for v in all_nodes:
                        if u != v and v in lengths_from_u:
                            fv = weight_n_t.get(v, 0)
                            # Add 1 to the distance to avoid division by zero for directly connected nodes with distance 0
                            sum_fv_div_duv += fv / (lengths_from_u[v] + 1)
                else:
                    # Handle isolated nodes or nodes not reachable from u
                    pass # Closeness for isolated nodes is typically 0 or undefined, handled by the division by total_weight_t


                current_closeness[u] = (fu + sum_fv_div_duv) / total_weight_t
        else:
             # If total node weight is 0 at time t, closeness is 0 for all nodes
             current_closeness = {node: 0 for node in all_nodes}


        # Sum up the closeness centrality for each node
        for node, closeness_value in current_closeness.items():
            temporal_closeness_sum[node] += closeness_value

    # Normalize the sum by the number of time steps in the interval
    normalized_temporal_closeness = {
        node: total_sum / num_time_steps for node, total_sum in temporal_closeness_sum.items()
    }

    return normalized_temporal_closeness, weight_n_t_history
```

```python

# Function to categorize highways based on their BR-xxx classification
def categorize_highway(highway_name):
    if pd.isna(highway_name):
        return "Unknown"
    # Assuming highway names are in the format 'BR-XXX'
    if isinstance(highway_name, str) and highway_name.startswith('BR-'):
        try:
            number_part = int(highway_name[3:])
            if 0 <= number_part <= 99:
                return "BR-0xx"
            elif 100 <= number_part <= 199:
                return "BR-1xx"
            elif 200 <= number_part <= 299:
                return "BR-2xx"
            elif 300 <= number_part <= 399:
                return "BR-3xx"
            elif 400 <= number_part <= 499:
                return "BR-4xx"
            # Add more categories as needed
            else:
                return "Other BR"
        except ValueError:
            return "Invalid BR Format"
    else:
        return "Not BR Highway"

highway_prefixes = ['BR-0', 'BR-1', 'BR-2', 'BR-3', 'BR-4', 'Other']
random.seed(42) # for reproducibility
dfg['Highway'] = dfg['geohash'].apply(lambda x: f"BR-{random.randint(0, 4)}{random.randint(0, 99)}")

# Apply the categorization function to the 'Highway' column
dfg['Highway_Category'] = dfg['Highway'].apply(categorize_highway)

print("\nNodes with Highway Categories:")
print(dfg[['geohash', 'Highway', 'Highway_Category']].head())


# Create a color map for highway categories
unique_categories = dfg['Highway_Category'].unique()
# Use a distinct colormap if available, or create a custom one
import matplotlib.cm as cm
colors = cm.get_cmap('tab10', len(unique_categories))
category_colors = {category: colors(i) for i, category in enumerate(unique_categories)}

# Map nodes to their highway category colors
node_category_colors = [category_colors.get(dfg[dfg['geohash'] == node]['Highway_Category'].iloc[0], 'gray') for node in graph.nodes()]

plt.figure(figsize=(12, 10))

# Draw the graph with nodes colored by highway category
pos = node_positions # Assuming node_positions is already defined

nx.draw(graph, pos, with_labels=False, node_size=50, node_color=node_category_colors, alpha=0.7, edge_color='lightblue')

plt.title("Highways in Northeastern Brazil by BR-xxx Category")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Create a legend for the highway categories
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=category,
                              markerfacecolor=color, markersize=10)
                   for category, color in category_colors.items()]
plt.legend(handles=legend_elements, title="Highway Category", loc='upper right')


plt.show()
```
