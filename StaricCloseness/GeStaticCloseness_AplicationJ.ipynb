```python

#DURO 1MIN
import pandas as pd

# Assuming the file is in the specified path
file_path = '/content/drive/MyDrive/DatosToUPE/todos_dataframes.xlsx'

try:
    # Read all sheets into a dictionary of DataFrames
    xls = pd.ExcelFile(file_path)
    all_dfs = {}
    for sheet_name in xls.sheet_names:
        all_dfs[sheet_name] = pd.read_excel(xls, sheet_name)

    graph_dict = {}

    for sheet_name, df in all_dfs.items():
        # Ensure the DataFrame is square (symmetric adjacency matrix)
        if df.shape[0] != df.shape[1]:
            print(f"Warning: DataFrame '{sheet_name}' is not square. Skipping.")
            continue

        # Set column names as index (row names)
        df.index = df.columns

        # Extract edges and weights
        edges_with_weights = {}
        for i in range(len(df.index)):
            for j in range(i + 1, len(df.columns)):  # Iterate through upper triangle
                weight = df.iloc[i, j]

                if weight != 0:
                    edges_with_weights[(df.index[i], df.columns[j])] = weight

        graph_dict[sheet_name] = edges_with_weights

except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")

# Now you have the graph data in the 'graph_dict'
# Example: Accessing data for the first sheet
# print(graph_dict[list(graph_dict.keys())[0]])
```

```python
# Assuming the file is in the specified path
file_path2 = '/content/drive/MyDrive/DatosToUPE/posicoes_geohash.xlsx'
try:
  dfg = pd.read_excel(file_path2)
  print(dfg.head()) # Display first few rows of the DataFrame
except FileNotFoundError:
  print(f"Error: File not found at {file_path2}")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python

import networkx as nx

# Create an empty graph
graph = nx.Graph()

# Add nodes with their geographical coordinates
for index, row in dfg.iterrows():
    geohash_index = row['geohash']  # Assuming 'geohash' column contains the node IDs
    longitude = row['longitude']
    latitude = row['latitude']
    graph.add_node(geohash_index, longitude=longitude, latitude=latitude)

# Add edges based on first_element.keys()
for edge in first_element.keys():
    source, target = edge
    weight = first_element[edge]
    if source in graph.nodes and target in graph.nodes:  # Check that nodes exists
        graph.add_edge(source, target, weight=weight)
    else:
        print(f"Warning: Edge ({source}, {target}) not added. One or both nodes are not in the graph.")
```

```python
import matplotlib.pyplot as plt
import networkx as nx

node_positions = {}
for node_data in dfg.to_dict('records'):
    node_id = node_data.get('geohash')
    longitude = node_data.get('longitude')
    latitude = node_data.get('latitude')
    if node_id and longitude and latitude:
        node_positions[node_id] = (longitude, latitude)

plt.figure(figsize=(10, 8))
nx.draw(graph, pos=node_positions, with_labels=False, node_size=20, node_color="skyblue")
plt.title(" Highways in Northeastern Brazil")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

# Static closeness centrality

```python


# Compute closeness centrality
closeness_centrality = nx.closeness_centrality(graph)

# Sort nodes by closeness centrality in descending order
sorted_closeness = sorted(closeness_centrality.items(), key=lambda item: item[1], reverse=True)

# Get the top 10 nodes
top_10_closeness = sorted_closeness[:10]

print("Top 10 nodes by Closeness Centrality:")
for node, centrality in top_10_closeness:
    print(f"Node: {node}, Closeness Centrality: {centrality:.6f}")
```

```python
top_10_node_ids = [node for node, centrality in top_10_closeness]

# Define node colors: red for top 10 nodes, lightblue otherwise
node_colors = []
for node in graph.nodes():
    if node in top_10_node_ids:
        node_colors.append('red')
    else:
        node_colors.append('lightblue')

# Plot the graph with highlighted nodes
plt.figure(figsize=(12, 10))
nx.draw(graph, pos=node_positions, with_labels=False, node_size=50, node_color=node_colors, alpha=0.7, edge_color='lightblue')
plt.title("Highways with Top 10 Central Nodes Highlighted (Closeness Centrality)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python


import matplotlib.patches as mpatches

# Calculate all four centrality metrics
degree_centrality = nx.degree_centrality(graph)
closeness_centrality = nx.closeness_centrality(graph)
betweenness_centrality = nx.betweenness_centrality(graph)
# Increase max_iter and/or change tol to improve convergence
eigenvector_centrality = nx.eigenvector_centrality(graph, max_iter=500, tol=1e-4)

# Store centrality values in dictionaries for easy lookup
centralities = {
    'degree': degree_centrality,
    'closeness': closeness_centrality,
    'betweenness': betweenness_centrality,
    'eigenvector': eigenvector_centrality
}

# Plot the graph with node colors based on which top 10 list they appear in
plt.figure(figsize=(12, 10))

# Create color mapping based on inclusion in top 10 for each centrality metric
node_colors_combined_all = []
top_10_degree_ids = [node for node, centrality in sorted(degree_centrality.items(), key=lambda item: item[1], reverse=True)[:10]]
top_10_closeness_ids = [node for node, centrality in sorted(closeness_centrality.items(), key=lambda item: item[1], reverse=True)[:10]]
top_10_betweenness_ids = [node for node, centrality in sorted(betweenness_centrality.items(), key=lambda item: item[1], reverse=True)[:10]]
top_10_eigenvector_ids = [node for node, centrality in sorted(eigenvector_centrality.items(), key=lambda item: item[1], reverse=True)[:10]]

for node in graph.nodes():
    in_degree = node in top_10_degree_ids
    in_closeness = node in top_10_closeness_ids
    in_betweenness = node in top_10_betweenness_ids
    in_eigenvector = node in top_10_eigenvector_ids

    if in_degree and in_closeness and in_betweenness and in_eigenvector:
        node_colors_combined_all.append('purple') # In all top 10s
    elif in_degree:
        node_colors_combined_all.append('red') # Only in top 10 degree
    elif in_closeness:
        node_colors_combined_all.append('blue') # Only in top 10 closeness
    elif in_betweenness:
        node_colors_combined_all.append('green') # Only in top 10 betweenness
    elif in_eigenvector:
        node_colors_combined_all.append('orange') # Only in top 10 eigenvector
    elif any([in_degree, in_closeness, in_betweenness, in_eigenvector]):
        node_colors_combined_all.append('yellow') # In top 10 of more than one (but not all)
    else:
        node_colors_combined_all.append('skyblue') # Not in any top 10


nx.draw(graph, pos=node_positions, with_labels=False, node_size=50, node_color=node_colors_combined_all, alpha=0.7, edge_color='lightblue')

# Create legend
#purple_patch = mpatches.Patch(color='purple', label='Top 10 in All')
red_patch = mpatches.Patch(color='red', label='Degree')
blue_patch = mpatches.Patch(color='blue', label='Closeness')
green_patch = mpatches.Patch(color='green', label='Betweenness')
orange_patch = mpatches.Patch(color='orange', label='Eigenvector')
#yellow_patch = mpatches.Patch(color='yellow', label='Top 10 in Multiple (not all)')
skyblue_patch = mpatches.Patch(color='skyblue', label='Others')

plt.legend(handles=[red_patch, blue_patch, green_patch, orange_patch, skyblue_patch], loc='upper left',fontsize='large')

#plt.title("Highways with Nodes Highlighted by Top 10 Centrality Rankings")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python


# Assuming the first column ('Unnamed: 0') contains the node IDs, set it as the index
df_rede = df_rede.set_index('Unnamed: 0')


# Create a list to store edges with weights from df_rede
rede_edges_with_weights = []

# Iterate through the DataFrame to extract edges and weights
for i in range(len(df_rede.index)):
    for j in range(len(df_rede.columns)):
        source_node = df_rede.index[i]
        target_node = df_rede.columns[j]
        weight = df_rede.iloc[i, j]

        # Add the edge with weight if the weight is not zero (assuming 0 means no connection)
        # You might want to adjust this condition based on what 0 represents in your data
        if weight != 0:
            rede_edges_with_weights.append((source_node, target_node, {'weight': weight}))

# Add these edges with weights to the existing 'graph'
# Use add_weighted_edges_from for convenience
graph.add_weighted_edges_from(rede_edges_with_weights)

print(f"Added {len(rede_edges_with_weights)} edges with weights from df_rede to the graph.")
print(f"Total number of edges in the graph: {graph.number_of_edges()}")

# Example: Print some of the edges with weights
# for u, v, data in list(graph.edges(data=True))[:10]:
#     print(f"Edge: ({u}, {v}), Weight: {data.get('weight')}")
```

```python

weighted_graph = nx.Graph()

# Add nodes from the original graph to the weighted graph to ensure all nodes are present.
# You might also want to copy node attributes if any were added to 'graph'.
for node, data in graph.nodes(data=True):
    weighted_graph.add_node(node, **data)

# Add the weighted edges from 'rede_edges_with_weights'
# The rede_edges_with_weights list contains tuples like (source, target, {'weight': weight})
# We need to ensure the weight is directly accessible as the 'weight' attribute for shortest path calculations.
weighted_graph.add_weighted_edges_from([(u, v, data['weight']) for u, v, data in rede_edges_with_weights])


print("Weighted graph 'weighted_graph' created.")
print(f"Number of nodes in weighted graph: {weighted_graph.number_of_nodes()}")
print(f"Number of edges in weighted graph: {weighted_graph.number_of_edges()}")

# You can verify by printing some edges with weights from the new graph
# for u, v, data in list(weighted_graph.edges(data=True))[:10]:
#     print(f"Weighted Edge: ({u}, {v}), Weight: {data.get('weight')}")
```

```python

weighted_closeness_centrality = nx.closeness_centrality(weighted_graph, distance='weight')

# Print the calculated weighted closeness centrality for each node
print("Weighted Closeness Centrality:")
for node, centrality in weighted_closeness_centrality.items():
    print(f"Node: {node}, Weighted Closeness Centrality: {centrality:.6f}")
```

```python
# Sort nodes by weighted closeness centrality in descending order
sorted_weighted_closeness = sorted(weighted_closeness_centrality.items(), key=lambda item: item[1], reverse=True)

# Get the top 10 nodes by weighted closeness centrality
top_10_weighted_closeness = sorted_weighted_closeness[:10]

print("\nTop 10 nodes by Weighted Closeness Centrality:")
for node, centrality in top_10_weighted_closeness:
    print(f"Node: {node}, Weighted Closeness Centrality: {centrality:.9f}")
```

```python
# Get the IDs of the top 10 nodes
top_10_weighted_node_ids = [node for node, centrality in top_10_weighted_closeness]

# Define node colors: red for top 10 nodes by weighted closeness, lightblue otherwise
node_colors_weighted = []
for node in weighted_graph.nodes():
    if node in top_10_weighted_node_ids:
        node_colors_weighted.append('red')
    else:
        node_colors_weighted.append('lightblue')

# Plot the weighted graph with highlighted nodes based on weighted closeness centrality
plt.figure(figsize=(12, 10))
# Use node_positions from the original graph as the spatial layout remains the same
nx.draw(weighted_graph, pos=node_positions, with_labels=False, node_size=50, node_color=node_colors_weighted, alpha=0.7, edge_color='lightblue')
plt.title("Highways with Top 10 Central Nodes Highlighted (Weighted Closeness Centrality)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Create a legend for the plot
red_patch_weighted = mpatches.Patch(color='red', label='Top 10 Weighted Closeness')
skyblue_patch_weighted = mpatches.Patch(color='lightblue', label='Others')
plt.legend(handles=[red_patch_weighted, skyblue_patch_weighted], loc='upper left', fontsize='large')

plt.show()
```
