```python

import pandas as pd

# Assuming the file is in the specified path
file_path = '/content/drive/MyDrive/DatosToUPE/todos_dataframes.xlsx'

try:
    # Read all sheets into a dictionary of DataFrames
    xls = pd.ExcelFile(file_path)
    all_dfs = {}
    for sheet_name in xls.sheet_names:
        all_dfs[sheet_name] = pd.read_excel(xls, sheet_name)

    graph_dict = {}

    for sheet_name, df in all_dfs.items():
        # Ensure the DataFrame is square (symmetric adjacency matrix)
        if df.shape[0] != df.shape[1]:
            print(f"Warning: DataFrame '{sheet_name}' is not square. Skipping.")
            continue

        # Set column names as index (row names)
        df.index = df.columns

        # Extract edges and weights
        edges_with_weights = {}
        for i in range(len(df.index)):
            for j in range(i + 1, len(df.columns)):  # Iterate through upper triangle
                weight = df.iloc[i, j]

                if weight != 0:
                    edges_with_weights[(df.index[i], df.columns[j])] = weight

        graph_dict[sheet_name] = edges_with_weights

except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import numpy as np
def calculate_transition_matrix(edges, edge_weights_t, q, num_nodes):
    """
    Calcula la matriz de probabilidad de transición para un instante de tiempo t.

    Args:
        edges (list of tuples): Lista de aristas, donde cada tupla es (origen, destino).
        edge_weights_t (dict): Diccionario donde las claves son las aristas (tuplas (origen, destino))
                               y los valores son sus pesos en el tiempo t.
        q (float): Parámetro q (0 <= q <= 1).
        num_nodes (int): Número total de nodos en la red.

    Returns:
        np.ndarray: Una matriz numpy de forma (num_nodes, num_nodes) que representa la
                      matriz de probabilidad de transición B(t).
    """
    transition_matrix = np.zeros((num_nodes, num_nodes))
    node_strengths = np.zeros(num_nodes)
    node_map = {node: i for i, node in enumerate(graph.nodes())}


    for (u, v), weight in edge_weights_t.items():
        u_mapped = node_map.get(u)
        if u_mapped is not None :
          node_strengths[u_mapped] += weight
        else:
            print(f"Warning: Node ID {u} not found in node map. Skipping this edge.")

    for i in range(num_nodes):
        si_t = node_strengths[i]
        for j in range(num_nodes):
            if si_t == 0:   #For each node i, consider a potential move to every node j (including itself).
                transition_matrix[i, j] = 1.0 if i == j else 0.0
            elif si_t >= 1:
                if i == j:
                    transition_matrix[i, j] = q ** si_t
                else:
                    i_original = list(graph.nodes())[i]
                    j_original = list(graph.nodes())[j]
                    weight_ij_t = edge_weights_t.get((i_original, j_original), 0) #Get the edge weight from node i to node j at time t, or 0 if no edge exists.
                    transition_matrix[i, j] = weight_ij_t * (1 - q ** si_t) / si_t #The remaining probability mass 1−𝑞𝑠𝑖(𝑡)1−qsi(t) is distributed proportionally to the edge weights from node i to its neighbors.

    return transition_matrix
```

### Dijkstra Temporal Probability

```python
import heapq

def dijkstra_temporal_probability(graph, start_node, end_node, transition_matrices):
    """
    Calculates the most probable path between two nodes in a time-varying graph using a Dijkstra-like algorithm.

    Args:
        graph: The graph represented as a NetworkX graph object.
        start_node: The starting node.
        end_node: The ending node.
        transition_matrices: A list of transition matrices, one for each time step.

    Returns:
        A tuple containing:
        - The probability of the most probable path.
        - The most probable path itself.
    """

    num_nodes = len(graph.nodes())
    num_time_steps = len(transition_matrices)
    node_map = {node: i for i, node in enumerate(graph.nodes())}  # Create a mapping from node ID to matrix index

    # Initialize distances and paths with probabilities
    distances = {node: 0 for node in graph.nodes()}  # Probability of reaching a node
    paths = {node: [] for node in graph.nodes()}
    distances[start_node] = 1  # Probability of starting at start_node is 1
    paths[start_node] = [start_node]

    priority_queue = [(1, start_node, 0)]  # (probability, node, time_step)

    while priority_queue:
        current_probability, current_node, current_time_step = heapq.heappop(priority_queue)

        if current_node == end_node:
            return current_probability, paths[end_node]

        next_time_step = (current_time_step + 1) % num_time_steps  # Move to the next time step (circular)
        transition_matrix = transition_matrices[next_time_step]

        # Iterate through neighbors in the graph, not matrix indices:
        for neighbor in graph.neighbors(current_node):
            # Get matrix indices for current_node and neighbor using the mapping:
            current_node_index = node_map.get(current_node)
            neighbor_index = node_map.get(neighbor)

            # Check if both nodes are in the mapping
            if current_node_index is not None and neighbor_index is not None:
                probability = transition_matrix[current_node_index, neighbor_index]
                new_probability = current_probability * probability

                if new_probability > distances[neighbor]:
                    distances[neighbor] = new_probability
                    paths[neighbor] = paths[current_node] + [neighbor]
                    heapq.heappush(priority_queue, (new_probability, neighbor, next_time_step))
            #else:
               # Handle the case where a node is not in the mapping (optional)
               # print(f"Warning: Node {current_node} or {neighbor} not found in the node mapping.")
    return 0, []  # No path found


def temporal_closeness_centrality_probability(graph, transition_matrices):
    """
    Computes the temporal closeness centrality based on probabilities for each node in the graph.

    Args:
        graph: A NetworkX graph object.
        transition_matrices: List of transition matrices for each time step.

    Returns:
        A dictionary mapping each node to its temporal closeness centrality (probability).
    """

    closeness_centralities = {}
    num_nodes = len(graph.nodes)
    for node in graph.nodes():
        total_probability = 0
        reachable_nodes = 0
        for target_node in graph.nodes():
            if node != target_node:
                probability, _ = dijkstra_temporal_probability(graph, node, target_node, transition_matrices)
                total_probability += probability
                if probability > 0:
                    reachable_nodes += 1  # Count only nodes with positive probability

        if reachable_nodes > 0:
            closeness_centralities[node] = total_probability / reachable_nodes
        else:
            closeness_centralities[node] = 0  # Handle cases with no paths

    return closeness_centralities
```

```python

temporal_closeness_centrality_probability = temporal_closeness_centrality_probability(graph, transition_matrices)

for node, centrality in temporal_closeness_centrality_probability.items():
    print(f"Node {node}: {centrality}")
```

```python
import heapq

def top_10_nodes(centrality_scores):
  """
  Returns the top 10 nodes with the highest centrality scores.

  Args:
    centrality_scores: A dictionary where keys are node IDs and values are centrality scores.

  Returns:
    A list of tuples, where each tuple contains a node ID and its centrality score, sorted in descending order of centrality.
  """
  return heapq.nlargest(10, centrality_scores.items(), key=lambda item: item[1])


top_nodes = top_10_nodes(temporal_closeness_centrality_probability)
print("Top 10 nodes with highest temporal closeness centrality:")
for node, centrality in top_nodes:
    print(f"Node {node}: {centrality}")
```

```python
import heapq
import math
import networkx as nx
import numpy as np

def dijkstra_temporal_log_distance(graph, start_node, end_node, transition_matrices):
    """
    Uses log-distance Dijkstra to find the most probable path in a temporal graph.

    Args:
        graph: networkx graph
        start_node: node where traversal begins
        end_node: node where traversal ends
        transition_matrices: list of 2D numpy arrays (probabilities) per time step

    Returns:
        A tuple: (max probability, path as list of nodes)
    """
    num_time_steps = len(transition_matrices)
    node_map = {node: i for i, node in enumerate(graph.nodes())}

    distances = {node: float('inf') for node in graph.nodes()}
    paths = {node: [] for node in graph.nodes()}
    distances[start_node] = 0
    paths[start_node] = [start_node]

    priority_queue = [(0, start_node, 0)]  # (log-cost, node, timestep)

    while priority_queue:
        current_cost, current_node, current_time = heapq.heappop(priority_queue)

        if current_node == end_node:
            max_probability = math.exp(-current_cost)
            return max_probability, paths[end_node]

        next_time = (current_time + 1) % num_time_steps
        transition_matrix = transition_matrices[next_time]

        for neighbor in graph.neighbors(current_node):
            i = node_map[current_node]
            j = node_map[neighbor]
            prob = transition_matrix[i, j]

            if prob > 0:
                edge_cost = -math.log(prob)
                new_cost = current_cost + edge_cost

                if new_cost < distances[neighbor]:
                    distances[neighbor] = new_cost
                    paths[neighbor] = paths[current_node] + [neighbor]
                    heapq.heappush(priority_queue, (new_cost, neighbor, next_time))

    return 0, []  # No path found


def temporal_closeness_centrality_probability_log(graph, transition_matrices):
    """
    Computes temporal closeness centrality for each node using the log-distance Dijkstra.

    Args:
        graph: networkx graph
        transition_matrices: list of 2D numpy arrays for each time step

    Returns:
        Dictionary of node closeness scores
    """
    closeness_centralities = {}

    for node in graph.nodes():
        total_probability = 0
        reachable_nodes = 0
        for target in graph.nodes():
            if node != target:
                prob, _ = dijkstra_temporal_log_distance(graph, node, target, transition_matrices)
                total_probability += prob
                if prob > 0:
                    reachable_nodes += 1

        if reachable_nodes > 0:
            closeness_centralities[node] = total_probability / reachable_nodes
        else:
            closeness_centralities[node] = 0

    return closeness_centralities
```
